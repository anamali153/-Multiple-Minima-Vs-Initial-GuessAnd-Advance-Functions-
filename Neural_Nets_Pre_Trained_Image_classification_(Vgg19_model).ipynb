create a function called format_img_vgg19() that takes file name as an argument. The function needs to be load the image in the default resolution for the VGG19 model, convert that image into an array and return the preprocessed image for the VGG19 model.

```python
def format_img_vgg19(filename):
  pic = load_img(filename, target_size=(224, 224))
  pic_arr = img_to_array(pic)
  expanded = np.expand_dims(pic_arr, axis=0)
  return preprocess_input(expanded)
```

#Load VGG19 Model

```python
%%time
vgg19_model = VGG19()
```

#Making Predictions

```python
data = format_img_vgg19(File_3)
prediction = vgg19_model.predict(data)
display(load_img(File_3))
decode_vgg19(prediction)
```
